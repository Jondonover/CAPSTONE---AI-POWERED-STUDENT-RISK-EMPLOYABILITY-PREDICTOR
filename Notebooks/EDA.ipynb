## 2 Loading datasets

```python
df_dropout  = pd.read_csv(PATH_DROPOUT, sep=';')
df_academic = pd.read_csv(PATH_ACADEMIC)
df_employ   = pd.read_csv(PATH_EMPLOY)


print("Shapes loaded:")
print(f"  df_dropout  : {df_dropout.shape}")
print(f"  df_academic : {df_academic.shape}")
print(f"  df_employ   : {df_employ.shape}")
```

```python
# Resume dataset (BOM + nested-CSV encoding + gzip compression)
def load_resume_csv(path):
    """Parse Resume.csv which wraps each row as a single quoted CSV cell and is gzip-compressed."""
    import gzip
    rows = []
    open_func = gzip.open if path.endswith('.gz') else open
    with open_func(path, 'rt', encoding='utf-8-sig') as f:
        for row in csv.reader(f):
            if row:
                rows.append(list(csv.reader([row[0]]))[0])
    header, data = rows[0], rows[1:]
    return pd.DataFrame(data, columns=header)

df_resume = load_resume_csv(PATH_RESUME)

print(f"Shape     : {df_resume.shape}")
print(f"Categories: {df_resume['category'].nunique()} unique")
df_resume.head(3)
```

## 3 Data cleaning

```python
#Standardizing column names
for df in [df_dropout, df_academic, df_employ, df_resume]:
    df.columns = (
        df.columns
        .str.strip()
        .str.lower()
        .str.replace(' ', '_')
        .str.replace('/', '_')
    )

print("Column names standardised")
```

```python
# removing duplicates
for name, df in [('df_dropout',  df_dropout),
                 ('df_academic', df_academic),
                 ('df_employ',   df_employ),
                 ('df_resume',   df_resume)]:
    before = len(df)
    df.drop_duplicates(inplace=True)
    print(f"{name}: {before} → {len(df)} rows ({before - len(df)} duplicates removed)")
```

```python
# handling missing data
for name, df in [('df_dropout',  df_dropout),
                 ('df_academic', df_academic),
                 ('df_employ',   df_employ),
                 ('df_resume',   df_resume)]:
    missing = df.isnull().sum()
    missing = missing[missing > 0]
    if missing.empty:
        print(f"{name}: no missing values ")
    else:
        print(f"\n{name} missing values:\n{missing}")

# Drop rows with nulls in critical columns (adjust as needed)
df_employ.dropna(inplace=True)
df_resume.dropna(inplace=True)
print("\n Missing values handled")
```

```python
# Map numeric codes to readable labels if not already strings
target_map = {0: 'Dropout', 1: 'Enrolled', 2: 'Graduate'}

if df_dropout['target'].dtype != object:
    df_dropout['target'] = df_dropout['target'].map(target_map)

print("Dropout target value counts:")
print(df_dropout['target'].value_counts())
```

### 7.3 Skill category scores

```python
score_col = 'employment_competitiveness_score'

# Compute correlation of every numeric feature with the target score
correlations = (
    df_employ
    .drop(columns=[score_col])
    .corrwith(df_employ[score_col])
    .dropna()
    .abs()
    .sort_values(ascending=True)
    .tail(8)
)

fig, ax = plt.subplots(figsize=(10, 5))
bars = ax.barh(
    correlations.index,
    correlations.values,
    color='#EF9F27',
    height=0.55,
    edgecolor='none'
)

for bar, val in zip(bars, correlations.values):
    ax.text(val + 0.005, bar.get_y() + bar.get_height() / 2,
            f'{val:.2f}', va='center', fontsize=10, color='#444')

ax.set_xlabel('Absolute correlation with Employment Competitiveness Score')
ax.set_title('Top 8 features correlated with employability score')
ax.set_xlim(0, correlations.max() * 1.15)
plt.tight_layout()
plt.savefig('outputs/chart9_feature_correlations.png', dpi=150, bbox_inches='tight')
plt.show()
```
